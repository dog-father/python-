格式正確的刻度線是準備發布的數據的重要組成部分。Matplotlib為滴答提供了一個完全可配置的系統。有刻度線定位器可以指定刻度線應出現的位置，刻度線格式化程序可以為刻度線提供所需的外觀。主刻度線和次刻度線可以相互獨立地定位和格式化。

# 問題: 繪製如下圖示:

 

```python
import matplotlib.pyplot as plt

#決定底框
plt.axes([0.1,0.1,.5,.5]) #[left, bottom, width, height]
#給定刻度
plt.xticks([]), plt.yticks([])
plt.text(0.1,0.1, 'axes([0.1,0.1,.5,.5])',ha='left',va='center',size=8,alpha=.5)

#help(plt.axes)

#決定第二層框
plt.axes([0.15,0.15,.4,.4])
#給定刻度
plt.xticks([]), plt.yticks([])
plt.text(0.15,0.15, 'axes([0.15,0.15,.4,.4])',ha='left',va='center',size=8,alpha=.5)

#決定第三層框
plt.axes([0.2,0.2,.3,.3])
#給定刻度
plt.xticks([]), plt.yticks([])
plt.text(0.2,0.2, 'axes([0.2,0.2,.3,.3])',ha='left',va='center',size=8,alpha=.5)

#決定第四層框
plt.axes([0.25,0.25,.2,.2])
#給定刻度
plt.xticks([]), plt.yticks([])
plt.text(0.25,0.25, 'axes([0.25,0.25,.2,.2])',ha='left',va='center',size=8,alpha=.5)

plt.show()
```

# 條型圖: Bar Plots
    
長條圖主要用來呈現兩個維度的資料，一個為X軸另一個則為Y軸(當然這邊指的是二維的狀況，較為常見)

主要用來呈現兩個維度的資料

# 問題: 嘗試通過添加紅色條形標籤重現右側的圖形。

```python
import numpy as np
import matplotlib.pyplot as plt

 #配置12 組 Bar
n = 12 
X = np.arange(n)
#print(X)
 #給定數學運算式
Y1 = (1-X/float(n)) * np.random.uniform(0.5,1.0,n)
Y2 = (1-X/float(n)) * np.random.uniform(0.5,1.0,n)
#指定上半部繪製區域, 給定 Bar 顏色, 邊界顏色
plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white')
# +Y 指的是 XY 四象限的第一象限

'''
#指定上半部繪製區域, 給定 Bar 顏色, 邊界顏色
'''
plt.bar(X, -Y2, facecolor='#998877', edgecolor='black')
print(X)
print('--------')
print(Y1)
print('--------')
ziped=zip(X,Y1)
print(list(ziped))
print('--------')
 #設定繪圖圖示區間
for x,y in zip(X,Y1):    
    plt.text(x+0.4, y+0.05, '%.2f' % y, ha='center', va= 'bottom')

for x,y2 in zip(X,Y2):    
    plt.text(x, -y2-0.15, '%.2f' % y2, ha='center', va= 'bottom')

 #設定Y軸區間
plt.ylim(-1.25,+1.25)
plt.show()
```
