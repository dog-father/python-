作業目標<br>
計算有缺失值的資料，統計量實作<br>
作業重點<br>
當遇到缺失值有函式可以處理，不須額外寫程式刪除<br>
計算統計量時不能出現缺失值

題目:<br>
english_score = np.array([55,89,76,65,48,70])<br>
math_score = np.array([60,85,60,68,np.nan,60])<br>
chinese_score = np.array([65,90,82,72,66,77])<br>
上3列共六位同學的英文、數學、國文成績，第一個元素代表第一位同學，舉例第一位同學英文55分、數學60分、國文65分，今天第五位同學因某原因沒來考試，導致數學成績缺值，運用上列數據回答下列問題。<br>
1. 請計算各科成績平均、最大值、最小值、標準差，其中數學缺一筆資料可忽略?
2. 第五位同學補考數學後成績為55，請計算補考後數學成績平均、最大值、最小值、標準差?
3. 用補考後資料找出與國文成績相關係數最高的學科?

```python
import numpy as np
```

```python
english_score = np.array([55,89,76,65,48,70])
math_score = np.array([60,85,60,68,np.nan,60])
chinese_score = np.array([65,90,82,72,66,77])
```

```python
#1. 請計算各科成績平均、最大值、最小值、標準差，其中數學缺一筆資料可忽略?
print('----english_score----')
print('mean:',np.nanmean(english_score))
print('max:',np.nanmax(english_score))
print('min:',np.nanmin(english_score))
print('std:',np.nanstd(english_score))
print('非nan筆數:',np.where((np.isnan(english_score)==False))[0].size)
print('----math_score----')
print('mean:',np.nanmean(math_score))
print('max:',np.nanmax(math_score))
print('min:',np.nanmin(math_score))
print('std:',np.nanstd(math_score))
print('非nan筆數:',np.where((np.isnan(math_score)==False))[0].size)
print('----chinese_score----')
print('mean:',np.nanmean(chinese_score))
print('max:',np.nanmax(chinese_score))
print('min:',np.nanmin(chinese_score))
print('std:',np.nanstd(chinese_score))
print('非nan筆數:',np.where((np.isnan(chinese_score)==False))[0].size)
```

```python
#2. 第五位同學補考數學後成績為55，請計算補考後數學成績平均、最大值、最小值、標準差?
math_score[4]=55
print(math_score)
print('----math_score----')
print('mean:',np.nanmean(math_score))
print('max:',np.nanmax(math_score))
print('min:',np.nanmin(math_score))
print('std:',np.nanstd(math_score))
print('非nan筆數:',np.where((np.isnan(math_score)==False))[0].size)
```

```python
#3. 用補考後資料找出與國文成績相關係數最高的學科?
print('國文成績:',chinese_score)
print('數學學成績:',math_score)
print('英文成績:',english_score)
print('--------')
#方法一
print('英文對數學的相關係數:',np.corrcoef(english_score,math_score))
print('國文對數學的相關係數:',np.corrcoef(chinese_score,math_score))
print('國文對英文的相關係數:',np.corrcoef(chinese_score,english_score))
```

```python
#方案二
import pandas as pd
a=np.array([english_score,math_score,chinese_score])
a=pd.DataFrame(a,index=['english', 'math', 'chinese'])
print(a.index)
print(a)
print('--------')
print(a.T)
print('--------')
print(a.T.corr())
```
