```python
#2.30分貝的聲壓會是50分貝的幾倍?
#公式移項過後可以得到 V1 = ?
#請寫下程式
#30分貝
V30=np.power(10,(30/20))*V0
print(V30)
#50分貝
V50=np.power(10,(50/20))*V0
print(V50)
print(V30/V50)
print(V50/V30) #50分貝是30分貝的10倍
```
