###  Q1: ÈÄèÈÅéÊï∏ÂÄºÊ≥ïË®àÁÆó  Age Âíå  Survived  ÊòØÂê¶ÊúâÁõ∏ÈóúÊÄß?


```python
#ÈÄ£Á∫åvs.Èõ¢Êï£
#Eta Squared(Œ∑2)
!pip install pingouin
import pingouin as pg
```

```python
#Step1: ÂèñÂá∫SurvivedÂíåAgeË≥áÊñô
#Step2:ÈÅãÁî® pg.anova Ë®àÁÆó‰∏âÁ®ÆËÆäÁï∞Êï∏
aov = pg.anova(dv='Age', between='SurvivedNew', data=df_train, detailed=True)
aov
```

```python
#Step3:ËÆäÁï∞Êï∏ÊèõÁÆóÂæóÂà∞ Eta Squared (ùúº^ùüê)
etaSq = aov.SS[0] / (aov.SS[0] + aov.SS[1])
etaSq
```

```python
def judgment_etaSq(etaSq):
    if etaSq < .01:
        qual = 'Negligible'
    elif etaSq < .06:
        qual = 'Small'
    elif etaSq < .14:
        qual = 'Medium'
    else:
        qual = 'Large'
    return(qual)
    
judgment_etaSq(etaSq)
```

```python
g = sns.catplot(x="SurvivedNew", y="Age", hue="SurvivedNew",
               data=df_train, kind="violin")
```

### Q2:ÈÄèÈÅéÊï∏ÂÄºÊ≥ïË®àÁÆó  Sex Âíå Survived  ÊòØÂê¶ÊúâÁõ∏ÈóúÊÄß?


```python
#Èõ¢Êï£vs.Èõ¢Êï£
#Cram√©r's V
!pip install researchpy

import researchpy   
```

```python
contTable = pd.crosstab(df_train['Age'], df_train['SurvivedNew'])
contTable
```

```python
df = min(df_train.shape[0], df_train.shape[1]) - 1
df
```

```python
df = min(contTable.shape[0], contTable.shape[1]) - 1
df #Ëá™Áî±Â∫¶
```

```python
crosstab, res = researchpy.crosstab(df_train['Age'], df_train['SurvivedNew'], test='chi-square')
print(res)
print("Cramer's value is",res.loc[2,'results'])
```

```python
## ÂØ´‰∏ÄÂÄãÂâØÁ®ãÂºèÂà§Êñ∑Áõ∏ÈóúÊÄßÁöÑÂº∑Â∫¶
def judgment_CramerV(df,V):
    if df == 1:
        if V < 0.10:
            qual = 'negligible'
        elif V < 0.30:
            qual = 'small'
        elif V < 0.50:
            qual = 'medium'
        else:
            qual = 'large'
    elif df == 2:
        if V < 0.07:
            qual = 'negligible'
        elif V < 0.21:
            qual = 'small'
        elif V < 0.35:
            qual = 'medium'
        else:
            qual = 'large'
    elif df == 3:
        if V < 0.06:
            qual = 'negligible'
        elif V < 0.17:
            qual = 'small'
        elif V < 0.29:
            qual = 'medium'
        else:
            qual = 'large'
    elif df == 4:
        if V < 0.05:
            qual = 'negligible'
        elif V < 0.15:
            qual = 'small'
        elif V < 0.25:
            qual = 'medium'
        else:
            qual = 'large'
    else:
        if V < 0.05:
            qual = 'negligible'
        elif V < 0.13:
            qual = 'small'
        elif V < 0.22:
            qual = 'medium'
        else:
            qual = 'large'
    return(qual)

judgment_CramerV(df,res.loc[2,'results'])
```
