題目:<br>
讀取STOCK_DAY_0050_202009.csv串聯STOCK_DAY_0050_202010.csv，並且找出open小於close的資料


```python
import pandas as pd
```

```python
#讀取STOCK_DAY_0050_202009.csv、STOCK_DAY_0050_202010.csv
a=pd.read_csv('STOCK_DAY_0050_202009.csv')
print(a)
print('----')
b=pd.read_csv('STOCK_DAY_0050_202010.csv')
print(b)
```

```python
#串聯
c=pd.concat([a,b],axis=0)
print(c)
```

```python
#找出open小於close的資料
print(c.loc[c.open<c.close])
```
